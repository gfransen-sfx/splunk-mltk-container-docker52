```python
# Step 1: Import the package
try:
    import networkx as nx
    import matplotlib.pyplot as plt
    print("Step 1: Importing networkx and matplotlib passed.")
except ImportError:
    print("Step 1 failed: 'networkx' or 'matplotlib' is not installed.")
    sys.exit(1)
except Exception as e:
    print(f"Step 1 failed with an unexpected error: {str(e)}")
    sys.exit(1)

# Step 2: Create a simple graph
try:
    G = nx.Graph()
    G.add_edge('A', 'B')
    G.add_edge('B', 'C')
    G.add_edge('C', 'A')
    print("Step 2: Creating a simple graph passed.")
except Exception as e:
    print(f"Step 2 failed: {str(e)}")
    sys.exit(1)

# Step 3: Compute basic graph properties
try:
    num_nodes = G.number_of_nodes()
    num_edges = G.number_of_edges()
    assert num_nodes == 3, f"Expected 3 nodes, got {num_nodes}"
    assert num_edges == 3, f"Expected 3 edges, got {num_edges}"
    print("Step 3: Computing basic graph properties passed.")
except Exception as e:
    print(f"Step 3 failed: {str(e)}")
    sys.exit(1)

# Step 4: Calculate shortest path
try:
    shortest_path = nx.shortest_path(G, source='A', target='C')
    expected_path = ['A', 'C']
    assert shortest_path == expected_path or shortest_path == ['A', 'B', 'C'], f"Expected {expected_path}, got {shortest_path}"
    print("Step 4: Calculating shortest path passed.")
except Exception as e:
    print(f"Step 4 failed: {str(e)}")
    sys.exit(1)

# Step 5: Draw the graph
try:
    plt.figure()
    nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    plt.title("Simple Graph")
    plt.show()
    print("Step 5: Drawing the graph passed.")
except Exception as e:
    print(f"Step 5 failed: {str(e)}")
    sys.exit(1)

# Final Confirmation
print("All extensive tests for the 'networkx' package completed successfully.")
```
